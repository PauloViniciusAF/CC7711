```python
import pandas as pd
import numpy as np
from sklearn import tree, metrics
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from scipy.io import arff
```

```python
data = pd.read_csv(".\jogadores.csv", sep = ',', encoding="UTF-8")

data_filtrado = data[data['Clube'] == 'Chapecoense'] 

data_filtrado
```

```python
preco = np.asarray(data_filtrado['Preço']).reshape(-1,1)
media= np.asarray(data_filtrado['Média']).reshape(-1,1)

features = np.concatenate((preco, media),axis=1)
target = data['resultado']
```

```python
Arvore = DecisionTreeClassifier(criterion='entropy').fit(features, target)

plt.figure(figsize=(10, 6.5))
tree.plot_tree(Arvore,feature_names=['Preço','Média', "G"],class_names=['Bom', 'Ruim'],
                   filled=True, rounded=True)
plt.show()

fig, ax = plt.subplots(figsize=(25, 10))
metrics.ConfusionMatrixDisplay.from_estimator(Arvore,features,target,display_labels=['Bom', 'Ruim'], values_format='d', ax=ax)
plt.show()
```
